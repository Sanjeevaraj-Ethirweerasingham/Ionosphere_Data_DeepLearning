Import Data

```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/Ionosphere.csv"
df = pd.read_csv(url)
df.head()
```

```python
#Assigning Variables

X = df.drop(["Class"], axis = 1)
y = df["Class"]

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

# Import Dependencies

```python
from keras.models import Sequential, load_model
from keras.layers import Dense,Dropout
```

# Build model

```python
model = Sequential()
model.add(Dense(units = 10,activation="relu",input_dim = len(X_train.columns)))
model.add(Dense(units=5,activation = "relu" ))
model.add(Dense(units=1,activation = "sigmoid"))
model.summary
```

```python
model.compile(loss = "binary_crossentropy", optimizer = "sgd", metrics = "accuracy")
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100, batch_size = 32)                         #epoch = number of times again and again data trained
```

```python
model.save("weights_h5")
```
